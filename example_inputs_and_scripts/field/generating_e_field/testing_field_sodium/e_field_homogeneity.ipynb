# Measuring the electric field homogeneity (positive x-direction field)

```python
import seaborn as sns

def set_style():
    # This sets reasonable defaults for font size for
    # a figure that will go in a paper
    sns.set_context("paper")
    
    # Set the font to be serif, rather than sans
    sns.set(font='serif')
    
    # Make the background white, and specify the
    # specific font family
    sns.set_style("white", {
        "font.family": "serif",
        "font.serif": ["Times", "Palatino", "serif"]
    })
    sns.despine()
    sns.set_style("ticks")

set_style()
```

Function to return the correct the x, y, or z, gradient of a singular Sodium ion from g.gradient obtained from ChemShell

```python
def return_gradient(file, d=''):
    #where 
    if d=='x':
        gradient = float((pd.read_csv(file, header=None))[0][4])
        
    if d=='y':
        gradient = float((pd.read_csv(file, header=None))[0][5])
    
    if d=='z':
        gradient = float((pd.read_csv(file, header=None))[0][6])
    return gradient
```

# The electric field in the pos-x direction when...
## Moving the sodium in the x, y, and, z direction

```python
# Moving sodium in the negative x-direction
files1 = sorted(glob.glob('9/x/neg/??/g.gradient'), reverse=True)
# Moving sodium in the positive x-direction
files2 = sorted(glob.glob('9/x/pos/??/g.gradient'))
# All g.gradient files correctly ordered
files = files1[:-1] +files2

x_grad = []
for file in files:
    x_grad.append(return_gradient(file, 'x'))

g = x_grad
x_perc = (np.array(g)/np.array(g[20])*100)-100

# Moving sodium in the negative y-direction
files1 = sorted(glob.glob('9/y/neg/??/g.gradient'), reverse=True)
# Moving sodium in the positive y-direction
files2 = sorted(glob.glob('9/y/pos/??/g.gradient'))
# All g.gradient files correctly ordered
files = files1[:-1] +files2

y_grad = []
for file in files:
    y_grad.append(return_gradient(file, 'x'))

g = y_grad
y_perc = (np.array(g)/np.array(g[20])*100)-100

# Moving sodium in the negative z-direction
files1 = sorted(glob.glob('9/z/neg/??/g.gradient'), reverse=True)
# Moving sodium in the positive z-direction
files2 = sorted(glob.glob('9/z/pos/??/g.gradient'))
# All g.gradient files correctly ordered
files = files1[:-1] +files2

z_grad = []
for file in files:
    z_grad.append(return_gradient(file, 'x'))

g = z_grad
z_perc = (np.array(g)/np.array(g[20])*100)-100
```

# How the positive-x electric field varies when a sodium is moved around a box

```python
plt.figure(figsize=(6,6),dpi=300)
plt.plot(np.arange(-20,21), x_perc, marker= 'o', alpha=0.8, label='x')
plt.plot(np.arange(-20,21), y_perc, marker= 'v', alpha=0.8, label='y')
plt.plot(np.arange(-20,21), z_perc, marker= '^', alpha=0.8, label='z')

plt.xlabel(r'Distance, $\mathrm{\AA}$')
plt.ylabel('Electric field divergence, %')
plt.legend()

plt.xticks()
plt.title('How the electric field applied in the +x direction diverges \n from the origin in the x, y and z directions')

sns.despine()
#plt.savefig('/Users/alex/Google Drive/PhD UCL/ImagesGraphs/e-field-divergence-x-BIG.png')
```
